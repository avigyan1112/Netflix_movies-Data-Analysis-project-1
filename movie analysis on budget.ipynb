```python
df_movieinfo = pd.read_csv('imdb.title.basics.csv.gz')
```

```python
df_movieinfo.rename(columns = {'primary_title':'movie'}, inplace = True)
```

```python
df_movieinfo
```

```python
df_merge_csv = pd.merge(df_movieinfo, df_tn_movie_budgets_csv, on = 'movie')
```

```python
df_merge_csv
```

```python
df_merge_csv['domestic_gross'] = df_merge_csv['domestic_gross'].str.replace('$','')
df_merge_csv['domestic_gross'] = df_merge_csv['domestic_gross'].str.replace(',','')
df_merge_csv['domestic_gross']
df_merge_csv
```

```python
  
import numpy as py
import matplotlib.pyplot as plt 
  
x = [1,2,3,4] 
y = [adven_val, Drama_val, comedy_val, action_val] 
labels = ['adven','drama', 'comedy','action']
plt.plot(x, y)
plt.title('Domestic_gross VS Gernes')
plt.xlabel('genres')
plt.ylabel('domestic_gross')
plt.xticks(x, labels, rotation ='vertical') 
 

plt.show()
```
